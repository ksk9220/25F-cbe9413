#### Task B

For this, we can use the standard formula to calculate the minimum work required. 

By setting the number of moles of CO2 (in output) as 1, we can calculate the relative moles of flue gas (with 4 mol CO2 content, and 90% efficiency), and the number of moles of exhaust. 

The formula to calculate the minimum work requirement (derived in class notes) is then given by - 

Wmin = -R*To(N1*sumfrom1tok(x1k*ln(x1k)) - N2*sumfrom1tok(x2k*ln(x2k)) - N3*sumfrom1tok(x3k*ln(x3k))) 

We consider flue gas as a binary mixture of CO2 and "rest of gas" (as given in lecture slides)


Now, setting alpha = 0.90(recovery)
x1,CO2 = 0.04
Hence, from the above formula -> N1 = 27.777

Again, using molar theory, since amount of CO2 remains conserved ideally, and since Product contains 90.6% of the CO2, the remaining 9.4% must be present in the residual gas.

Conserving CO2 – 

0.94*0.04*N1 = (N1-1)* x2,CO2

LHS - 9.4% of the C02 originally present. The C02 originally present is N1* x1,CO2, and x1,CO2 = 0.04.

RHS – Moles of N3 present = N1-1. Together, (N1-1)* x2,CO2 depicts the moles of CO2 in the residual gas.

Solving, we get x2,CO2 =0.0039.

Also, x1,everything else =0.96
Also, x2,everything else =0.996

Now, we can use the formula mentioned above to calculate the minimum work requirement.

Solving, Wmin = 9799.56 J/mol of CO2 captured.



```python
import math

## Assumtion : unit of N = mol/s
N3 = 1
alpha = 0.906
x1co2 = 0.04

# From the formula given above - 

N1 = 1/(alpha*x1co2)
N2 = (N1-1)

x2co2 = N1*x1co2*0.094/(N2)

x1else = 1 - x1co2
x2else = 1 - x2co2

print(f"N1 = {N1}, N2 = {N2}, x2co2 = {x2co2}, x2else = {x2else}")

```

```python
# Calculating Wmin
R = 8.314  # J/(mol·K)
T = 298    # K (25°C)
Wmin = -1* R * T * (N1*(x1co2*math.log(x1co2) + x1else*math.log(x1else)) - N2*(x2co2*math.log(x2co2) + x2else*math.log(x2else)) - N3*(1*math.log(1)))

print(f"Wmin = {Wmin:.2f} J/mol of CO2 captured")
```
